```python
X = df.drop(columns=['species'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(X_scaled)
```

```python
df['cluster'] = kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score
```

```python
df['KMeans_cluster']=kmeans.fit_predict(X_scaled)
```

```python
silhouette_avg = silhouette_score(X_scaled, df['KMeans_cluster'])
print(f'silhouette score :{silhouette_avg:.2f}')
```

```python
# Visualize the clusters based on the first two features (sepal length and sepal width)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['KMeans_cluster'], cmap='viridis', marker='o')
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('KMeans Clustering of Iris Dataset')
plt.xlabel(df.columns[0])  # Feature 1 (sepal length)
plt.ylabel(df.columns[1])  # Feature 2 (sepal width)
plt.legend()
plt.show()
```
